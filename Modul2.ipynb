
- Inaccurate Values: Inaccurate values occur when the values in a dataset do not match the observation results.

- Outliers:Outliers, also known as anomalies, are data points that are significantly different from other data points in a dataset. These significantly different values will affect several statistical parameters used to analyze data, such as mean and standard deviation.

There are several methods that can be used to identify outliers in a dataset. The most commonly used method is the IQR (Interquartile Range) method. IQR identifies outliers by creating cut-off values as a factor (usually we use a value from 1.5 to 3) of the IQR value. These cut-off values are then used to calculate boundary values. The values are divided into two: minimum and maximum thresholds. All data points smaller than the minimum threshold or larger than the maximum threshold are considered outliers.


```python
import numpy as np
import seaborn as sns

data = [6, 2, 3, 4, 5, 1, 50]
sort_data = np.sort(data)
sort_data

# Percentile : The value that divides the data into one hundred equal parts.
Q1 = np.percentile(data, 25, interpolation = 'midpoint')
Q2 = np.percentile(data, 50, interpolation = 'midpoint')
Q3 = np.percentile(data, 75, interpolation = 'midpoint')

print('Q1 25 percentile of the given data is, ', Q1)
print('Q1 50 percentile of the given data is, ', Q2)
print('Q1 75 percentile of the given data is, ', Q3)

IQR = Q3 - Q1
print('Interquartile range is', IQR)
```

```python
# Find the lower and upper limits as Q1 ‚Äì 1.5 IQR and Q3 + 1.5 IQR, respectively.
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
```

```python
# Data points greater than the upper limit or less than the lower limit are outliers
outlier =[]
for x in data:
    if ((x> up_lim) or (x<low_lim)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
```

```python
# Plot the box plot to highlight outliers.
sns.boxplot(data)
```

```python
# Calculate IQR
from scipy import stats
IQR = stats.iqr(data, interpolation = 'midpoint')
IQR
```

#### üçÄ Techniques for Handling Outliers

a. Dropping : The first method, which is the easiest, is to drop or remove all rows containing outliers.

```python
import pandas as pd

df = pd.read_csv("data.csv")

Q1 = df['TotalCharges'].quantile(0.25)
Q3 = df['TotalCharges'].quantile(0.75)
IQR = Q3 - Q1

maximum = Q3 + (1.5*IQR)
minimum = Q1 - (1.5*IQR)

condition_lower_than = df['TotalCharges'] < minimum
condition_more_than = df['TotalCharges'] > maximum

df.drop(df[condition_lower_than].index, inplace=True)
df.drop(df[condition_more_than].index, inplace=True)
```

b. Imputation : The concept is similar to the previous method, which is replacing outliers with certain values. The values we can use are the mean, median, mode, or boundary values.

```python
import pandas as pd
df = pd.read_csv("data.csv")

Q1 = df['TotalCharges'].quantile(0.25)
Q3 = df['TotalCharges'].quantile(0.75)
IQR = Q3 - Q1

maximum = Q3 + (1.5*IQR)
minimum = Q1 - (1.5*IQR)

condition_lower_than = df['TotalCharges'] < minimum
condition_more_than = df['TotalCharges'] > maximum

df.mask(cond=condition_more_than, maximum, axis=1, inplace=True)
df.mask(cond=condition_lower_than, minimum, axis=1, inplace=True)
```
